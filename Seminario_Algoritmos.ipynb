Se trata de un problema de **maximización**, ya que lo que buscamos es conseguir precisamente la mayor cantidad de audiencia total en millones de espectadores.

La **función objetivo** será la encargada de calcular cuánta audiencia genera un calendario completo propuesto. Para esto, sumamos la audiencia de cada partido calculada así: multiplicar la audiencia base por la ponderación del horario, y luego le restamos si hay partidos cruzados a la misma hora (coincidencias) como una forma de penalización:

$\displaystyle \text{Max} \sum_{i=1}^{10} \Big( \text{AudienciaBase}(p_i) \times \text{Ponderacion}(h_i) \times (1 - \text{Penalizacion}(C_{h_i})) \Big) $

Donde $C_{h_i}$ representa la cantidad de partidos extra que cayeron en esa misma franja horaria. De igual forma, para garantizar que haya partidos viernes y lunes, también se aplicaría una penalización fuerte si se detecta que en uno de estos dos días no hay partidos asignados. Lo que decidimos es multiplicar la audiencia total por 0.001 como penalización para estos casos. De ese modo, si una solución tiene un valor potencialmente alto, pero no cumple con dicha restricción, su valor total se verá muy afectado con la penalización. 

**Diseña un algoritmo para resolver el problema por fuerza bruta**

_Respuesta_

```python
# Datos base del problema
audiencia_base = {
    ('A', 'A'): 2.0, ('A', 'B'): 1.3, ('A', 'C'): 1.0,
    ('B', 'A'): 1.3, ('B', 'B'): 0.9, ('B', 'C'): 0.75,
    ('C', 'A'): 1.0, ('C', 'B'): 0.75, ('C', 'C'): 0.47
}

# 10 franjas: 0=V20, 1=S12, 2=S16, 3=S18, 4=S20, 5=D12, 6=D16, 7=D18, 8=D20, 9=L20
ponderacion = [0.4, 0.55, 0.7, 0.8, 1.0, 0.45, 0.75, 0.85, 1.0, 0.4]

# Porcentaje de audiencia que se mantiene según coincidencias (1 partido = 100% de audiencia)
# Si hay 2 partidos (1 coincidencia) se pierde el 25%, por lo que se mantiene el 0.75, etc.
remanente_coincidencia = [1.0, 0.75, 0.55, 0.40, 0.30, 0.25, 0.22, 0.20, 0.20, 0.20]

# Lista de partidos de prueba según las categorías de los equipos enfrentados
# Ejemplo: ('B', 'A') representa un equipo de categoría B contra uno de categoría A
partidos_ejemplo = [
    ('B', 'A'), ('B', 'A'), ('C', 'C'), ('B', 'A'), ('C', 'C'),
    ('B', 'C'), ('B', 'B'), ('B', 'B'), ('B', 'C'), ('B', 'B')
]

# Genera de forma recursiva todas las configuraciones posibles de horarios.
def explorador_combinaciones(opciones, profundidad):
    if profundidad == 0:
        yield []
    else:
        for opcion in opciones:
            for combinacion_restante in explorador_combinaciones(opciones, profundidad - 1):
                yield [opcion] + combinacion_restante

# El Juez de Audiencia que audita la rentabilidad de cada calendario propuesto
def evaluar_calendario(solucion, partidos):
    audiencia_total = 0
    coincidencias = [0] * 10
    
    # Contamos rapido cuantos partidos hay en cada franja
    for franja in solucion:
        coincidencias[franja] += 1
        
    # Calculo interdependiente de la funcion objetivo
    for i in range(len(solucion)):
        franja = solucion[i]
        eq1, eq2 = partidos[i]
        base = audiencia_base[(eq1, eq2)]
        pond = ponderacion[franja]
        
        # Num de coincidencias es el total de partidos en la franja menos el actual
        num_coin = coincidencias[franja] - 1
        rem = remanente_coincidencia[num_coin]
        
        audiencia_total += base * pond * rem
        
    # Restricción obligatoria: Al menos un partido el Viernes (0) y uno el Lunes (9)
    if coincidencias[0] == 0 or coincidencias[9] == 0:
        audiencia_total *= 0.001
        
    return audiencia_total

def fuerza_bruta_la_liga(partidos):
    mejor_solucion = None
    mejor_audiencia = -1
    iteraciones = 0
    
    # Generamos y evaluamos cada escenario posible de horarios para los 10 partidos
    for solucion in explorador_combinaciones(range(10), len(partidos)):
        audiencia = evaluar_calendario(solucion, partidos)
        
        if audiencia > mejor_audiencia:
            mejor_audiencia = audiencia
            mejor_solucion = solucion
            
        iteraciones += 1
        
        # Rompemos el bucle prematuramente para no normalizar la alienacion del hardware.
        # Calcular los 10 mil millones colapsaria el notebook humano.
        if iteraciones > 5000000:
            print("Se aplica una salida de emergencia para no colapsar los recursos del equipo.")
            break
            
    return mejor_solucion, mejor_audiencia

# Prueba del algoritmo de fuerza bruta
bf_sol, bf_aud = fuerza_bruta_la_liga(partidos_ejemplo)
print("Mejor hallado por fuerza bruta parcial:", bf_sol, "Audiencia:", bf_aud)
```

**Calcula la complejidad del algoritmo por fuerza bruta**

_Respuesta_

La complejidad de probar todas las combinaciones mediante fuerza bruta es de $\mathcal{O}(H^N)$, donde $H$ sería la cantidad de franjas horarias disponibles (10) y $N$ sería la cantidad de partidos (10).

**(*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta**

_Respuesta_

En principio pensamos en usar algoritmos voraces, pero el problema es que si el algoritmo toma siempre la decisión que da más audiencia inmediata, terminaría metiendo todos los mejores partidos el sábado a las 20h y la penalización por coincidencia nos terminaría corrompiendo el total. También descartamos de entrada la búsqueda local simple porque pensamos que sería muy fácil que el algoritmo se estancara en un óptimo local, porque tras cualquier cambio pequeño se podría empeorar el resultado actual.

Así que optamos por algoritmos genéticos, porque en lugar de ir a ciegas probando todos los posibles escenarios, mantiene una población de posibles calendarios y va cruzando los que dan mejor audiencia. También pensamos que un punto clave aquí sería la mutación. Al cambiar partidos de horario al azar, forzamos al algoritmo a explorar otras configuraciones y salir de posibles bloqueos. 

```python
import random

# Se genera una solucion aleatoria basica
def crear_solucion():
    # Retorna un calendario aleatorio de 10 partidos en 10 franjas
    return [random.randint(0, 9) for _ in range(10)]

# Genera una poblacion inicial de soluciones de tamano N
def generar_poblacion(N):
    poblacion = []
    for _ in range(N):
        poblacion.append(crear_solucion())
    return poblacion

# Evalua la poblacion y devuelve el mejor individuo
def Evaluar_Poblacion(poblacion, partidos):
    mejor_solucion = []
    mejor_audiencia = -1 # Inicializamos en negativo para maximizar
    
    for p in poblacion:
        audiencia_actual = evaluar_calendario(p, partidos) # Reutilizamos el fitness de arriba
        if audiencia_actual > mejor_audiencia:
            mejor_solucion = p
            mejor_audiencia = audiencia_actual
            
    return mejor_solucion, mejor_audiencia

# Funcion de mutacion. Se elige un partido al azar y se cambia su franja horaria
def Mutar(solucion, mutacion):
    # Forzamos un cambio estocastico para evadir los optimos locales
    if random.random() < mutacion:
        i = random.randint(0, len(solucion) - 1)
        solucion[i] = random.randint(0, 9)
    return solucion

# Funcion para generar hijos a partir de 2 padres
def Descendencia(padres, mutacion_prob):
    padre1, padre2 = padres
    # Punto de corte aleatorio
    corte = random.randint(1, len(padre1) - 1)
    
    # Cruzamos el material genetico
    hijo1 = padre1[:corte] + padre2[corte:]
    hijo2 = padre2[:corte] + padre1[corte:]
    
    # A diferencia del problema TSP, en La Liga SI podemos repetir franjas
    # por lo que evitamos la vieja funcion 'Factibilizar'.
    
    return [Mutar(hijo1, mutacion_prob), Mutar(hijo2, mutacion_prob)]

# Funcion de cruce. Recibe una poblacion y devuelve la poblacion ampliada
def Cruzar(poblacion, mutacion_prob):
    poblacion_copia = [solucion[:] for solucion in poblacion]
    poblacion_final = [solucion[:] for solucion in poblacion]
    
    while len(poblacion_copia) > 1:
        # Extraemos dos padres aleatorios de la poblacion
        padre1 = poblacion_copia.pop(random.randrange(len(poblacion_copia)))
        padre2 = poblacion_copia.pop(random.randrange(len(poblacion_copia)))
        
        # Generamos los hijos y los integramos
        hijos = Descendencia([padre1, padre2], mutacion_prob)
        poblacion_final.extend(hijos)
        
    return poblacion_final

# Funcion de seleccion de la poblacion con elitismo protector
def Seleccionar(poblacion, partidos, N, elitismo):
    # Evaluamos toda la sociedad
    pob_evaluada = [(sol, evaluar_calendario(sol, partidos)) for sol in poblacion]
    
    # Ordenamos de mayor a menor audiencia
    pob_evaluada.sort(key=lambda x: x[1], reverse=True)
    
    # Protegemos inmaculada a la elite
    n_elite = int(N * elitismo)
    elite = [sol for sol, aud in pob_evaluada[:n_elite]]
    
    # Del resto, seleccionamos organicamente para no perder diversidad genetica
    n_restantes = N - n_elite
    restantes = [sol for sol, aud in pob_evaluada[n_elite:]]
    seleccionados = random.sample(restantes, min(n_restantes, len(restantes)))
    
    return elite + seleccionados

# Funcion principal del algoritmo genetico
def algoritmo_genetico(partidos, N=100, mutacion=0.15, elitismo=0.1, generaciones=100):
    # Genera la poblacion inicial
    poblacion = generar_poblacion(N)
    
    # Inicializamos valores para la mejor solucion
    mejor_solucion, mejor_audiencia = Evaluar_Poblacion(poblacion, partidos)
    
    parar = False
    n = 0
    
    # Iniciamos el ciclo evolutivo
    while(parar == False):
        # Cruce y mutacion
        poblacion = Cruzar(poblacion, mutacion)
        # Seleccion de la nueva poblacion
        poblacion = Seleccionar(poblacion, partidos, N, elitismo)
        # Evaluamos la nueva sociedad
        mejor_solucion, mejor_audiencia = Evaluar_Poblacion(poblacion, partidos)
        
        if n % 40 == 0:
            print(f"Generacion #{n} \nLa mejor solucion es: {mejor_solucion} \ncon audiencia {mejor_audiencia:.3f} Millones\n")
            
        if n == generaciones:
            parar = True
        n += 1
        
    print(f"Generacion FINAL #{n-1} \nLa mejor solucion dictaminada es: {mejor_solucion} \ncon audiencia maxima de {mejor_audiencia:.3f} Millones\n")
    return mejor_solucion, mejor_audiencia
```

**(*)Calcula la complejidad del algoritmo**

_Respuesta_

La complejidad de esta propuesta la calculamos revisando los ciclos principales de nuestra implementación, donde el tiempo es $\mathcal{O}(G \cdot P \cdot N)$
Donde $G$ son las generaciones iteradas, $P$ el tamaño de la poblacion, y $N$ la cantidad de partidos a evaluar en el juez de audiencia. El proceso de seleccion introduce un $\mathcal{O}(P \log P)$ por el ordenamiento, pero en resumen, es una complejidad lineal controlada y sostenida respecto a $N$. 

En espacio ocupa $\mathcal{O}(P \cdot N)$ porque solo necesitamos guardar las listas de la población en cada generación.
