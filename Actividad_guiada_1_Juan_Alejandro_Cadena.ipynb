## Torres de Hanoi con Divide y vencerás

```python
def Torres_Hanoi(N, desde, hasta):
  if N ==1 :
    print("Lleva la ficha " ,desde , " hasta ", hasta )

  else:
    #Torres_Hanoi(N-1, desde, 6-desde-hasta )
    Torres_Hanoi(N-1, desde, 6-desde-hasta )
    print("Lleva la ficha " ,desde , " hasta ", hasta )
    #Torres_Hanoi(N-1,6-desde-hasta, hasta )
    Torres_Hanoi(N-1, 6-desde-hasta  , hasta )

Torres_Hanoi(3, 1 , 3)
```

## Sucesión_de_Fibonacci
- https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci
- Calculo del termino n-simo de la suscesión de Fibonacci

```python
def Fibonacci(N:int):
  if N < 2:
    return 1
  else:
    return Fibonacci(N-1)+Fibonacci(N-2)

Fibonacci(5)
```

**Mi análisis**:
* **Tiempo:** $\mathcal{O}(2^n)$. Esta recursión recalcula varias veces los mismos subproblemas, por lo que aún podría ser más eficiente. 
* **Espacio:** $\mathcal{O}(n)$ por la pila de recursión en memoria.

Para evitar ese recálculo se podrían guardar los dos valores calculados previamente, lo que reduciría el tiempo a **$\mathcal{O}(n)$** y el espacio a **$\mathcal{O}(1)$**. A continuación la mejora propuesta:

```python
def fibonacci_optimizado(N: int) -> int:
    if N < 2:
        return 1
    
    # Memoria O(1): Solo necesitamos recordar los dos últimos valores calculados
    a = 1 # Fibonacci(n-2)
    b = 1 # Fibonacci(n-1)
    
    for _ in range(2, N + 1):
        siguiente_valor = a + b
        a = b
        b = siguiente_valor
        
    return b

Fibonacci(5)
```

## Devolución de cambio por técnica voraz

```python
def cambio_monedas(N, SM):
  SOLUCION = [0]*len(SM)   #SOLUCION = [0,0,0,0,..]
  ValorAcumulado = 0

  for i,valor in enumerate(SM):
    monedas =  (N-ValorAcumulado)//valor
    SOLUCION[i] = monedas
    ValorAcumulado = ValorAcumulado + monedas*valor

    if ValorAcumulado == N:
      return SOLUCION


cambio_monedas(15,[25,10,5,1])
```

**Mi análisis**
* **Tiempo:** $\mathcal{O}(m)$, donde $m$ es el número de monedas, pues sólo se está recorriendo la lista una vez (el resto son operaciones constantes, como divisiones, multiplicaciones y sumas)
* **Espacio:** $\mathcal{O}(m)$ para almacenar el arreglo de resultados.

## N-Reinas por técnica de vueta atrás


```python
def escribe(S):
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")


def es_prometedora(SOLUCION,etapa):
  #print(SOLUCION)
  #Si la solución tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " está " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True



def reinas(N, solucion=[], etapa=0):
  if len(solucion) == 0:
      solucion=[0 for i in range(N)]

  for i in range(1, N+1):
    solucion[etapa] = i

    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
        #escribe(solucion)
        print()
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(8)
```

**Mi análisis**

* **Tiempo:** $\mathcal{O}(N!)$. El árbol de ramificación de opciones es factorial siendo esta la función dominante.
* **Espacio:** $\mathcal{O}(N)$ por el arreglo en memoria y la pila de recursión.


## Viaje por el rio. Programación dinámica

```python
TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]



################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #Inicialización de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################


PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```

**Mi análisis**
* **Tiempo:** $\mathcal{O}(N^3)$. Al utilizar tres ciclos anidados (i, j, k) estilo Floyd-Warshall, calcula la distancia mínima posible entre cualquier par de puntos, pero esto puede hacer que la solución resulte pesada.
* **Espacio:** $\mathcal{O}(N^2)$ al inicializar matrices completas $N \times N$.

Si sólo buscamos la ruta del origen al destino (del nodo 0 al N-1), se podría simplificar utilizando arreglos de 1 dimensión de izquierda a derecha, de ese modo el tiempo bajaría a **$\mathcal{O}(N^2)$** y el espacio a **$\mathcal{O}(N)$**. A continuación la mejora propuesta:

```python
def viaje_rio_optimizado(TARIFAS):
    N = len(TARIFAS)
    
    # coste_minimo[i] guardará el viaje más barato desde el origen(0) hasta el nodo 'i'.
    # Inicializamos con "infinito", excepto el origen porque de 0 a 0 el coste es 0.
    coste_minimo = [float('inf')] * N
    coste_minimo[0] = 0
    
    # predecesor[i] guardará desde qué embarcadero saltamos a 'i' de forma más barata.
    # Es el "rastreador" fundamental para poder reconstruir la ruta física al final.
    predecesor = [-1] * N
    
    # Este ciclo evalúa cada embarcadero destino posible (desde el 1 hasta el N-1)
    for destino in range(1, N):
        # Este ciclo evalúa cada posible punto intermedio (origen) desde el cual saltar al destino
        for origen in range(destino):
            tarifa = TARIFAS[origen][destino]
            if tarifa < 999:
                if coste_minimo[origen] + tarifa < coste_minimo[destino]:
                    coste_minimo[destino] = coste_minimo[origen] + tarifa
                    predecesor[destino] = origen
                
    # Ahora se reconstruye la ruta leyendo el la traza hacia atrás
    ruta_optima = []
    nodo_actual = N - 1
    
    while nodo_actual != -1:
        ruta_optima.append(nodo_actual)
        nodo_actual = predecesor[nodo_actual]
        
    ruta_optima.reverse()
    ruta_formateada = ",".join(map(str, ruta_optima))
    print(f"La ruta es:\n{ruta_formateada}")
    
    return coste_minimo[N-1], ruta_optima

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```
